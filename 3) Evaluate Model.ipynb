# Importing Libraries

```python
from keras.models import model_from_json
import cv2, numpy as np,os
```

# Loading Model

```python
PATH = "3-class-model/"
json_file = open(PATH + "model.json", "r")
model_json = json_file.read()
json_file.close()
model = model_from_json(model_json)
# load weights into new model
model.load_weights(PATH + "model.h5")
print("Loaded model from disk")
```

```python
labels_name = ['Fighting', 'Normal', 'Robbery']
```

```python
def plot_n_predict(img):
    # plt.imshow(img)
    img = cv2.resize(img, (64, 64))
    img = img.reshape(1, 64, 64, 3)
    pred = model.predict(img)
    pred_label = labels_name[np.argmax(pred)]
    # plt.title("Predicted = {}".format(pred_label))
    # plt.show()
    return np.argmax(pred)
```

```python
def get_probability(val, val_list):
    return val_list.count(val)/len(val_list) * 100
```

```python
# predicting
video_path = 'videos/Normal'
full_path = os.path.join(video_path, os.listdir(video_path)[8])
full_path = 'fight.mp4'
```

```python
cap = cv2.VideoCapture(full_path)
predictions = []
i = 0
print('Predicting ...', end='')
while cap.isOpened():
    ret, frame = cap.read()
    if ret == False:
        break
    pred = plot_n_predict(frame)
    predictions.append(pred)
    i+=1
    
unique_list = list(set(predictions))
for val in unique_list:
    prob = round(get_probability(val, predictions),2 )
    print("\nVideo contains {} of {}% probability.".format(labels_name[val], prob), end='')
```

```python
# walk through every frames
for i in range(len(predictions)):
    print("At {} frames, Video contains {} scene.".format(i, labels_name[predictions[i]]))
```
